### Cost: 
Determines the accuracy of our model to predict the target price of the house
$$J(w,b)= \frac{1}{2m}\sum\limits_{i=0}^{m-1} (f_{w,b}(x^{(i)})-(y^{(i)}))^2$$
where 
$$f_{w,b}(x^{(i)}) = wx^{(i)}+b $$
> Note: In code, the summation will be from 0 to m-1 (arrays in python).
- Parameters - w and b
- $(f_{w,b}(x^{(i)}) - y^{(i)} )^2$ is the squared difference between teh target value and the prediction
- Divided the sum by '2m' to poduce a much neater result

```python
def compute_cost(x,y,w,b):
    m = len(x)
    cost_sum = 0
    for i in range(m):
        f_wb=w*x[i]+b
        cost = (f_wb - y[i])**2
        cost_sum = cost+cost_sum
        
    total_cost = (1/(2*m))*cost_Sum
    return total_cost
```

## Cost Function Inituition
$f_{w,b}(x^{(i)}) = wx^{(i)}+b $
We can predict house values using this model by changing the values of w and b.
- Input x: size of the house
- Cost Function: to measure of how accurate the model is on the training data
- w and b should be selected such that the squared error is 0, minimizing the cost function


- The cost is minimised when w = 200, b is not considered for the sake of simplicity
- Since the difference between the target and the predicted values are squared, they might increase rapidly
- Since, the model forms a straight line, the coefficients are a perfect fit to the data
